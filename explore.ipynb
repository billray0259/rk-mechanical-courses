```python
import pandas as pd

```

```python
file1 = 'data/combined.xlsx'
df = pd.read_excel(file1, sheet_name="2019-2020")
df
```

```python
df.dropna(inplace=True, how='all')
df.to_csv("course_catalogue.csv", index=False)
```

```python
df["Course #"].nunique()
```

```python
grouped = df.groupby("Course #")
singles = []
for course_num, group in grouped:
    if len(group) > 1:
        print(f"Course #: {course_num}")
        display(group)
    else:
        singles.append(course_num)
        
print(f"Single course numbers: {singles}")
```
